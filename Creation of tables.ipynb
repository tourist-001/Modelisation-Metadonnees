```python
import psycopg2 as psy

# Connection
connection  = psy.connect( user = "postgres", password = "password", host= "localhost", port="5432")
cur = connection.cursor()


# Creation des tables

# yellow
# mot "User" a été réservé dans la base de données on utilise donc Userr
user = """
        CREATE TABLE Userr (
            id SERIAL PRIMARY KEY,
               lastName VARCHAR(255) NOT NULL,
               firstName VARCHAR(255) NOT NULL,
               privilege VARCHAR(255) NOT NULL
        )
        """

jobtitle = """CREATE TABLE JobTitle(
                   id SERIAL PRIMARY KEY,
                   description VARCHAR(255) NOT NULL,
                   jobTitle VARCHAR(255) NOT NULL)"""

hasjobtitle = """CREATE TABLE hasJobTitle(
                      userid INTEGER NOT NULL,
                      jobtitleid INTEGER NOT NULL,
                      FOREIGN KEY (userid)
                              REFERENCES Userr(id)
                              ON UPDATE CASCADE ON DELETE CASCADE,
                      FOREIGN KEY (jobtitleid)
                              REFERENCES JobTitle(id)
                              ON UPDATE CASCADE ON DELETE CASCADE,
                      PRIMARY KEY (userid, jobtitleid))"""

#DataSet
dataset = """CREATE TABLE Dataset(
                  id SERIAL PRIMARY KEY,
                  name VARCHAR(255) NOT NULL,
                  type VARCHAR(255) NOT NULL,
                  location VARCHAR(255) NOT NULL,
                  creationDate DATE NOT NULL)"""

#orange
datasetsource = """CREATE TABLE DatasetSource(
                        id INTEGER PRIMARY KEY,
                        FOREIGN KEY (id)
                                REFERENCES Dataset(id)
                                ON UPDATE CASCADE ON DELETE CASCADE,
                        owner VARCHAR(255) NOT NULL)"""

sourceofsteam ="""CREATE TABLE SourceOfSteam(
                       id SERIAL PRIMARY KEY,
                       description VARCHAR(255) NOT NULL,
                       sourceid INTEGER NOT NULL,
                       FOREIGN KEY (sourceid)
                               REFERENCES DatasetSource(id)
                               ON UPDATE CASCADE ON DELETE CASCADE)"""

#purple
sensitivitylevel = """CREATE TABLE SensitivityLevel(
                           level INTEGER PRIMARY KEY,
                           description VARCHAR(500) NOT NULL)"""

sensitivitymark = """CREATE TABLE SensitivityMark(
                          id SERIAL PRIMARY KEY,
                          description VARCHAR(255) NOT NULL,
                          sensitivitylevel INTEGER NOT NULL,
                          FOREIGN KEY (sensitivitylevel)
                                  REFERENCES Sensitivitylevel(level)
                                  ON UPDATE SET NULL ON DELETE SET NULL)"""

#DataLakeDataSet
datalakedataset = """CREATE TABLE DatalakeDataset(
                         id INTEGER PRIMARY KEY,
                         description VARCHAR(255) NOT NULL,
                         connectionURL VARCHAR(255) NOT NULL,
                         size VARCHAR(255) NOT NULL,
                         administrator VARCHAR(255) NOT NULL,
                         filenameExtension VARCHAR(255) NOT NULL,
                         sensitivitymarkid INTEGER,
                         FOREIGN KEY (id)
                                 REFERENCES Dataset(id)
                                 ON UPDATE CASCADE ON DELETE CASCADE,
                         FOREIGN KEY (sensitivitymarkid)
                                 REFERENCES SensitivityMark(id)
                                 ON UPDATE SET NULL ON DELETE SET NULL)"""

#Ingest
ingest = """CREATE TABLE Ingest(
                 id SERIAL NOT NULL,
                 ingestionMode VARCHAR(255) NOT NULL,
                 ingestionStartTime VARCHAR(255) NOT NULL,
                 definedDuration INTEGER NOT NULL,
                 ingetionEndTime VARCHAR(255) NOT NULL,
                 ingestionMethodeName VARCHAR(255) NOT NULL,
                 ingestionSourceCodeUrl VARCHAR(255) NOT NULL,
                 ingestionBinaryMachineCodeUrl VARCHAR(255) NOT NULL,
                 ingestionEnvironement VARCHAR(255) NOT NULL,
                 ingestionOutputLog VARCHAR(255) NOT NULL,
                 ingestionErrorLog VARCHAR(255) NOT NULL,
                 ingestionComment VARCHAR(255) NOT NULL,
                 userid INTEGER NOT NULL,
                 datalakedatasetid INTEGER NOT NULL,
                 datasetsourceid INTEGER NOT NULL,
                 FOREIGN KEY (userid)
                         REFERENCES Userr(id)
                         ON UPDATE CASCADE ON DELETE CASCADE,
                 FOREIGN KEY (datalakedatasetid)
                         REFERENCES DatalakeDataset(id)
                         ON UPDATE CASCADE ON DELETE CASCADE,
                 FOREIGN KEY (datasetsourceid)
                         REFERENCES DatasetSource(id)
                         ON UPDATE CASCADE ON DELETE CASCADE,
                 PRIMARY KEY (id, userid, datalakedatasetid, datasetsourceid))"""

#blue
qualitydimension = """CREATE TABLE QualityDimension(
                           id SERIAL PRIMARY KEY,
                           name VARCHAR(255) NOT NULL,
                           description VARCHAR(255) NOT NULL)"""

qualitymetric = """CREATE TABLE QualityMetric(
                         id SERIAL PRIMARY KEY,
                         name VARCHAR(255) NOT NULL,
                         unit VARCHAR(255) NOT NULL,
                         dimentionid INTEGER NOT NULL,
                         FOREIGN KEY (dimentionid)
                                 REFERENCES QualityDimension(id)
                                 ON UPDATE CASCADE ON DELETE CASCADE)"""

qualitymeasurement = """CREATE TABLE QualityMeasurement(
                             name VARCHAR(255) NOT NULL,
                             formula VARCHAR(255) NOT NULL,
                             value NUMERIC NOT NULL,
                             datalakedatasetid INTEGER NOT NULL,
                             qualitymetricid INTEGER NOT NULL,
                             FOREIGN KEY (datalakedatasetid)
                                     REFERENCES DatalakeDataset(id)
                                     ON UPDATE CASCADE ON DELETE CASCADE,
                             FOREIGN KEY (qualitymetricid)
                                     REFERENCES QualityMetric(id)
                                     ON UPDATE CASCADE ON DELETE CASCADE,
                             PRIMARY KEY (datalakedatasetid, qualitymetricid))"""

#red
globaldict = """CREATE TABLE GlobalDict(
                     id SERIAL PRIMARY KEY,
                     name VARCHAR(255) NOT NULL,
                     description VARCHAR(255) NOT NULL)"""

hasglobaldict = """CREATE TABLE hasGlobalDict(
                        dictid INTEGER NOT NULL,
                        datalakedatasetid INTEGER NOT NULL,
                        FOREIGN KEY (dictid)
                                REFERENCES GlobalDict(id)
                                ON UPDATE CASCADE ON DELETE CASCADE,
                        FOREIGN KEY (datalakedatasetid)
                                REFERENCES DatalakeDataset(id)
                                ON UPDATE CASCADE ON DELETE CASCADE,
                        PRIMARY KEY (dictid, datalakedatasetid))"""

#pink
relationshipds = """CREATE TABLE RelationshipDS(
                         id SERIAL PRIMARY KEY,
                         name VARCHAR(255) NOT NULL,
                         description VARCHAR(255) NOT NULL)"""

analysisdsrelationship = """CREATE TABLE AnalysisDSRelationship(
                                 name VARCHAR(255) NOT NULL,
                                 description VARCHAR(255) NOT NULL,
                                 value NUMERIC NOT NULL,
                                 relationshipid INTEGER NOT NULL,
                                 datalakedatasetid1 INTEGER NOT NULL,
                                 datalakedatasetid2 INTEGER NOT NULL,
                                 FOREIGN KEY (relationshipid)
                                         REFERENCES RelationshipDS
                                         ON UPDATE CASCADE ON DELETE CASCADE,
                                 FOREIGN KEY (datalakedatasetid1)
                                         REFERENCES DatalakeDataset(id)
                                         ON UPDATE CASCADE ON DELETE CASCADE,
                                 FOREIGN KEY (datalakedatasetid2)
                                         REFERENCES DatalakeDataset(id)
                                         ON UPDATE CASCADE ON DELETE CASCADE,
                                 PRIMARY KEY (relationshipid, datalakedatasetid1,datalakedatasetid2))"""

#green
tag = """CREATE TABLE Tag(
              id SERIAL PRIMARY KEY,
              name VARCHAR(255) NOT NULL)"""

hastag = """CREATE TABLE hasTag(
                 tagid INTEGER NOT NULL,
                 datalakedatasetid INTEGER NOT NULL,
                 FOREIGN KEY (tagid)
                         REFERENCES Tag(id)
                         ON UPDATE CASCADE ON DELETE CASCADE,
                 FOREIGN KEY (datalakedatasetid)
                                         REFERENCES DatalakeDataset(id)
                                         ON UPDATE CASCADE ON DELETE CASCADE,
                 PRIMARY KEY(tagid, datalakedatasetid))"""

dlunstructureddataset = """CREATE TABLE DLUnstructuredDataset(
                                datalakedatasetid INTEGER PRIMARY KEY,
                                format VARCHAR(255) NOT NULL,
                                FOREIGN KEY (datalakedatasetid)
                                         REFERENCES DatalakeDataset(id)
                                         ON UPDATE CASCADE ON DELETE CASCADE)"""

dlsemistructureddataset = """CREATE TABLE DLSemiStructuredDataset(
                                  datalakedatasetid INTEGER PRIMARY KEY,
                                  nbentityclasses INTEGER,
                                  FOREIGN KEY (datalakedatasetid)
                                         REFERENCES DatalakeDataset(id)
                                         ON UPDATE CASCADE ON DELETE CASCADE)"""

dlstructureddataset = """CREATE TABLE DLStructuredDataset(
                              datalakedatasetid INTEGER PRIMARY KEY,
                              nbentityclasses INTEGER,
                              FOREIGN KEY (datalakedatasetid)
                                         REFERENCES DatalakeDataset(id)
                                         ON UPDATE CASCADE ON DELETE CASCADE)"""

entityclass = """CREATE TABLE EntityClass(
                      id SERIAL PRIMARY KEY,
                      name VARCHAR(255) NOT NULL,
                      comment VARCHAR(255) NOT NULL,
                      semistructuredid INTEGER,
                      structuredid INTEGER,
                      FOREIGN KEY (semistructuredid)
                              REFERENCES DLSemiStructuredDataset(datalakedatasetid)
                              ON UPDATE CASCADE ON DELETE CASCADE,
                      FOREIGN KEY (structuredid)
                              REFERENCES DLStructuredDataset(datalakedatasetid)
                              ON UPDATE CASCADE ON DELETE CASCADE,
                      sensitivitymarkid INTEGER,
                      FOREIGN KEY (sensitivitymarkid)
                              REFERENCES SensitivityMark(id)
                              ON UPDATE SET NULL ON DELETE SET NULL)"""

attribute = """CREATE TABLE Attribute(
                    id SERIAL NOT NULL,
                    name VARCHAR(255) NOT NULL,
                    type VARCHAR(255) NOT NULL,
                    entityclassid INTEGER NOT NULL,
                    sensitivitymarkid INTEGER,
                    FOREIGN KEY (entityclassid)
                            REFERENCES EntityClass(id)
                            ON UPDATE CASCADE ON DELETE CASCADE,
                    FOREIGN KEY (sensitivitymarkid)
                            REFERENCES SensitivityMark(id)
                            ON UPDATE SET NULL ON DELETE SET NULL)"""

# executer les creations
cur.execute(user)
cur.execute(jobtitle)
cur.execute(hasjobtitle)
cur.execute(dataset)
cur.execute(datasetsource)
cur.execute(sourceofsteam)
cur.execute(sensitivitylevel)
cur.execute(sensitivitymark)
cur.execute(datalakedataset)
cur.execute(ingest)
cur.execute(qualitydimension)
cur.execute(qualitymetric)
cur.execute(qualitymeasurement)
cur.execute(globaldict)
cur.execute(hasglobaldict)
cur.execute(relationshipds)
cur.execute(analysisdsrelationship)
cur.execute(tag)
cur.execute(hastag)
cur.execute(dlunstructureddataset)
cur.execute(dlsemistructureddataset)
cur.execute(dlstructureddataset)
cur.execute(entityclass)
cur.execute(attribute)

# enregistrer les changements
connection.commit()

# fermer la connection
connection.close()
```
